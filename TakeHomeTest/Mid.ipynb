```python
import pandas_datareader.data as web 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.pylab as pylab

## Prob 1 
data = pd.read_csv('NBOption.csv') 
plt.hist(data['Volatility'],bins = 20, alpha = 0.3,label = 'Volatility') 
plt.title('Volatility & Implied Volatility of Option') 
plt.ylabel('count')
plt.hist(data['ImpliedVolatility'], bins = 20, alpha = 0.3, label = 'Implied Volatility') 
plt.legend(loc = 'upper right')
plt.show()
```

```python
x_train = np.array(data['Volatility']) 
print(x_train) 
y_train = np.array(data['ImpliedVolatility']) 
print(y_train) 
x = np.array ([0.284653603,1.24219129]) 
plt.scatter(x_train,y_train,color = 'g') 
plt.scatter(x[0],x[1],color = 'b') 
plt.show()

distances = [] 
for x_train in x_train:

    d = (np.sum((x_train - x[0])**2+(y_train - x[1])**2))**0.5
    distances.append(d)

nearest = np.argsort(distances) 
k = 25 
NearestK_y = [y_train[i] for i in nearest [:k]] 
print(NearestK_y)
```

#### The KNN could be used to predict volatility and applied volatility, because the volatility could be repeat or forecast from historic data, and the forecast data could be find around the point be given.
